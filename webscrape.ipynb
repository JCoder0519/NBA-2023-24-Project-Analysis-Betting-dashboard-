```python
import pandas as pd
from nba_api.stats.endpoints import shotchartdetail
from nba_api.stats.static import players
import time

# Function to get all active players
def get_all_active_players():
    return players.get_active_players()

# Function to fetch shot data for a player
def fetch_shot_data(player_id, season='2023-24', season_type='Regular Season'):
    try:
        shot_data = shotchartdetail.ShotChartDetail(
            team_id=0,
            player_id=player_id,
            season_nullable=season,
            season_type_all_star=season_type,
            context_measure_simple='FGA'
        ).get_data_frames()
        return shot_data[0] if len(shot_data) > 0 else pd.DataFrame()
    except Exception as e:
        print(f"Error fetching data for player ID {player_id}: {e}")
        return pd.DataFrame()

# Get all active players
all_players = get_all_active_players()

# Initialize an empty DataFrame to store all shot data
all_shots_df = pd.DataFrame()

# Loop through each player and fetch their shot data
for player in all_players:
    player_id = player['id']
    player_name = player['full_name']
    print(f"Fetching data for {player_name} (ID: {player_id})")

    # Fetch shot data for the current player
    player_shots_df = fetch_shot_data(player_id)

    # Add player name to the DataFrame
    if not player_shots_df.empty:
        player_shots_df['player_name'] = player_name
        # Append to the main DataFrame
        all_shots_df = pd.concat([all_shots_df, player_shots_df], ignore_index=True)

    # Sleep to avoid hitting rate limits
    time.sleep(1)  # Adjust as needed

# Save the combined data to a CSV file
all_shots_df.to_csv('all_players_shot_data_2023_2024.csv', index=False)
print("All players' shot data saved to all_players_shot_data_2023_2024.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# List of NBA team abbreviations
teams = [
    'ATL', 'BOS', 'BRK', 'CHO', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW',
    'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK',
    'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS'
]

base_url = "https://www.basketball-reference.com/teams/{}/2024_games.html"

all_game_logs = []

def get_team_game_logs(team):
    url = base_url.format(team)
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to load page {url}")
        return []

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the game log table
    table = soup.find('table', {'id': 'games'})
    if not table:
        print(f"No game log table found for {team}")
        return []

    headers = [th.getText() for th in table.find('thead').findAll('th')]
    rows = table.find('tbody').findAll('tr')
    
    team_logs = []
    for row in rows:
        if row.find('th', {"scope": "row"}):
            game_log = {header: cell.getText() for header, cell in zip(headers, row.findAll(['th', 'td']))}
            team_logs.append(game_log)

    return team_logs

for team in teams:
    team_logs = get_team_game_logs(team)
    all_game_logs.extend(team_logs)
    print(f"Processed team {team}")
    time.sleep(1)  # To avoid overwhelming the server

# Convert to DataFrame
game_logs_df = pd.DataFrame(all_game_logs)

# Save to CSV
game_logs_df.to_csv('nba_team_game_logs_2023_2024.csv', index=False)
print("Game logs saved to nba_team_game_logs_2023_2024.csv")
```

```python
from nba_api.stats.endpoints import leaguegamefinder, boxscoresummaryv2
import pandas as pd
import time

# Function to fetch all games for the 2023-2024 season
def fetch_nba_games_2023_2024():
    gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable='2023-24')
    games = gamefinder.get_data_frames()[0]
    return games

# Function to fetch box score summary for a specific game
def fetch_box_score_summary(game_id):
    boxscore = boxscoresummaryv2.BoxScoreSummaryV2(game_id=game_id)
    return boxscore

# Fetching games
games_2023_2024 = fetch_nba_games_2023_2024()

# Initialize list to store line scores
line_scores_list = []

new_dataframe = pd.DataFrame()

# Fetch line scores for each game
for game_id in games_2023_2024['GAME_ID'].unique():
    box_score_summary = fetch_box_score_summary(game_id)
    line_score = box_score_summary.line_score.get_data_frame()
    game_info = box_score_summary.game_summary.get_data_frame()
    
    if not game_info.empty:
        game_date = game_info['GAME_DATE_EST'].values[0]
        home_team_id = game_info['HOME_TEAM_ID'].values[0]
        visitor_team_id = game_info['VISITOR_TEAM_ID'].values[0]
        
        for index, row in line_score.iterrows():
            line_scores_list.append({
                'GAME_ID': game_id,
                'GAME_DATE': game_date,
                'HOME_TEAM_ID': home_team_id,
                'VISITOR_TEAM_ID': visitor_team_id,
                'TEAM_ID': row['TEAM_ID'],
                'TEAM_ABBREVIATION': row['TEAM_ABBREVIATION'],
                'PTS_QTR1': row['PTS_QTR1'],
                'PTS_QTR2': row['PTS_QTR2'],
                'PTS_QTR3': row['PTS_QTR3'],
                'PTS_QTR4': row['PTS_QTR4']
            })
            new_dataframe = pd.concat([new_dataframe, pd.DataFrame(line_scores_list[-1], index=[0])], ignore_index=True)
            print(line_scores_list[-1])
    else:
        print(f"No game summary available for GAME_ID: {game_id}")
        
    time.sleep(0.1)  # Sleep to avoid hitting the API rate limit
    
new_dataframe.to_csv('nba_game_line_scores_2023_2024.csv', index=False)
```

```python
from nba_api.stats.endpoints import teamgamelog, commonteamyears, teaminfocommon
import pandas as pd

# Function to fetch all NBA teams for the current season
def fetch_all_teams():
    teams = commonteamyears.CommonTeamYears()
    teams_df = teams.get_data_frames()[0]
    return teams_df['TEAM_ID'].unique()

# Function to fetch game logs for a specific team
def fetch_team_game_log(team_id, season='2023-24'):
    game_log = teamgamelog.TeamGameLog(team_id=team_id, season=season)
    game_log_df = game_log.get_data_frames()[0]
    return game_log_df[['Game_ID', 'GAME_DATE', 'MATCHUP', 'PTS', 'WL']]

# Function to fetch team info to get team names
def fetch_team_info(team_id):
    team_info = teaminfocommon.TeamInfoCommon(team_id=team_id)
    team_info_df = team_info.get_data_frames()[0]
    return team_info_df['TEAM_NAME'].values[0]

# Fetching all team IDs
team_ids = fetch_all_teams()

# Initialize list to store all teams' game logs
all_teams_game_logs = []

# Fetch game logs for each team
for team_id in team_ids:
    team_game_log = fetch_team_game_log(team_id)
    team_name = fetch_team_info(team_id)
    team_game_log['TEAM_NAME'] = team_name
    all_teams_game_logs.append(team_game_log)

# Concatenate all team game logs into a single DataFrame
all_teams_game_logs_df = pd.concat(all_teams_game_logs, ignore_index=True)

# Add a column for the winning team name
all_teams_game_logs_df['WINNING_TEAM'] = all_teams_game_logs_df.apply(
    lambda row: row['TEAM_NAME'] if row['WL'] == 'W' else row['MATCHUP'].split()[-1],
    axis=1
)

# Exporting the dataframe to a CSV file
csv_file_path = 'nba_teams_game_logs_2023_2024.csv'
all_teams_game_logs_df.to_csv(csv_file_path, index=False)

print(f"NBA teams' game logs for the 2023-2024 season have been exported to {csv_file_path}")
```

```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd

# Function to fetch NBA game matchups for the 2023-2024 season
def fetch_nba_game_matchups_2023_2024():
    game_finder = leaguegamefinder.LeagueGameFinder(
        season_nullable='2023-24',
        season_type_nullable= 'Regular Season'
    )
    games_df = game_finder.get_data_frames()[0]
    
    # Filter and rename columns
    relevant_columns = {
        'GAME_ID': 'GAME_ID',
        'TEAM_ID': 'HOME_TEAM_ID',
        'TEAM_NAME': 'HOME_TEAM_NAME',
        'TEAM_ABBREVIATION': 'HOME_TEAM_ABBR',
        'MATCHUP': 'MATCHUP'
    }
    
    games_df = games_df.rename(columns=relevant_columns)[list(relevant_columns.values())]

    # Extract home and away team information from the 'MATCHUP' column
    games_df['AWAY_TEAM_NAME'] = games_df['MATCHUP'].apply(lambda x: x.split(' ')[0])
    games_df['HOME_TEAM_NAME'] = games_df['MATCHUP'].apply(lambda x: x.split(' ')[2])
    games_df = games_df[['GAME_ID', 'HOME_TEAM_NAME', 'AWAY_TEAM_NAME']]

    return games_df

# Fetching game matchups
game_matchups_2023_2024 = fetch_nba_game_matchups_2023_2024()

# Exporting the dataframe to a CSV file
csv_file_path = 'nba_game_matchups_2023_2024.csv'
game_matchups_2023_2024.to_csv(csv_file_path, index=False)

print(f"NBA game matchups for the 2023-2024 season have been exported to {csv_file_path}")
```

```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd

# Function to fetch NBA game matchups for the 2023-2024 season
def fetch_nba_game_matchups_2023_2024():
    game_finder = leaguegamefinder.LeagueGameFinder(
        season_nullable='2023-24',
        season_type_nullable= 'Playoffs'
    )
    games_df = game_finder.get_data_frames()[0]
    
    # Filter and rename columns
    relevant_columns = {
        'GAME_ID': 'GAME_ID',
        'TEAM_ID': 'HOME_TEAM_ID',
        'TEAM_NAME': 'HOME_TEAM_NAME',
        'TEAM_ABBREVIATION': 'HOME_TEAM_ABBR',
        'MATCHUP': 'MATCHUP'
    }
    
    games_df = games_df.rename(columns=relevant_columns)[list(relevant_columns.values())]

    # Extract home and away team information from the 'MATCHUP' column
    games_df['AWAY_TEAM_NAME'] = games_df['MATCHUP'].apply(lambda x: x.split(' ')[0])
    games_df['HOME_TEAM_NAME'] = games_df['MATCHUP'].apply(lambda x: x.split(' ')[2])
    games_df = games_df[['GAME_ID', 'HOME_TEAM_NAME', 'AWAY_TEAM_NAME']]

    return games_df

# Fetching game matchups
game_matchups_2023_2024 = fetch_nba_game_matchups_2023_2024()

# Exporting the dataframe to a CSV file
csv_file_path = 'playoff_matchups_2023_2024.csv'
game_matchups_2023_2024.to_csv(csv_file_path, index=False)

print(f"NBA game matchups for the 2023-2024 season have been exported to {csv_file_path}")
```

```python
from nba_api.stats.endpoints import commonallplayers, commonplayerinfo
import pandas as pd

# Fetch all active players
players = commonallplayers.CommonAllPlayers(is_only_current_season=1)
players_data = players.get_data_frames()[0]

# Filter to keep only relevant data
players_filtered = players_data[['PERSON_ID', 'DISPLAY_FIRST_LAST']]

# Initialize an empty list to store player positions
player_positions = []

# Fetch position for each player using commonplayerinfo
for player_id in players_filtered['PERSON_ID']:
    profile = commonplayerinfo.CommonPlayerInfo(player_id=player_id)
    player_info = profile.get_normalized_dict()
    if player_info and 'CommonPlayerInfo' in player_info:
        position = player_info['CommonPlayerInfo'][0]['POSITION']
    else:
        position = 'Unknown'
    player_positions.append(position)

# Add positions to the DataFrame
players_filtered['POSITION'] = player_positions

# Display the DataFrame
print(players_filtered)

# Optionally, save the DataFrame to a CSV file
players_filtered.to_csv('active_nba_players_with_positions.csv', index=False)
```

```python
import requests
import pandas as pd
import time

# Constants
HEADERS = {
    "Host": "stats.nba.com",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    "Accept": "application/json, text/plain, */*",
    "Referer": "https://stats.nba.com/",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-US,en;q=0.9",
}

# Function to get all active player IDs for a specific season
def get_active_player_ids(season):
    url = "https://stats.nba.com/stats/commonallplayers"
    params = {
        "IsOnlyCurrentSeason": "0",
        "LeagueID": "00",
        "Season": season,
    }
    response = requests.get(url, headers=HEADERS, params=params)
    data = response.json()
    players = data["resultSets"][0]["rowSet"]
    # Filter to get only active players (IsActive column is at index 3 in the rowSet)
    active_players = {player[0]: player[2] for player in players if player[3] == 1}
    return active_players

# Function to get game logs for a specific player and season
def get_player_game_logs(player_id, season):
    url = "https://stats.nba.com/stats/playergamelogs"
    params = {
        "PlayerID": player_id,
        "Season": season,
        "SeasonType": "Regular Season"
    }
    response = requests.get(url, headers=HEADERS, params=params)
    data = response.json()
    headers = data["resultSets"][0]["headers"]
    rows = data["resultSets"][0]["rowSet"]
    return pd.DataFrame(rows, columns=headers)

# Main script to get game logs for all active players for the 2022-2023 season
all_game_logs = pd.DataFrame()
season = "2021-22"
player_ids = get_active_player_ids(season)
for i, (player_id, player_name) in enumerate(player_ids.items()):
    print(f"Fetching game logs for {player_name} ({i+1}/{len(player_ids)}) in season {season}")
    player_logs = get_player_game_logs(player_id, season)
    if not player_logs.empty:
        player_logs["PlayerID"] = player_id
        player_logs["PlayerName"] = player_name
        player_logs["Season"] = season
        all_game_logs = pd.concat([all_game_logs, player_logs], ignore_index=True)
    time.sleep(1)  # To avoid overwhelming the server

# Save to CSV
if not all_game_logs.empty:
    all_game_logs.to_csv('nba_player_game_logs_active_players_2021_22.csv', index=False)
    print("Game logs saved to nba_player_game_logs_active_players_2021_22.csv")
else:
    print("No game logs found for any active players in the specified season.")
```
